## 📦 Part 1 — Variables

A **variable** is like a labelled box — you store a value inside and give it a name.
Python works out whether it's a number, text, or something else automatically.

```python
# Storing numbers and text
age  = 11
name = "Theo"
pi   = 3.14159

print(f"Hello {name}, you are {age} years old")
print(f"Pi is roughly {pi}")
```

```python
# Maths with variables
radius = 5
area   = pi * radius ** 2
circ   = 2 * pi * radius

print(f"Circle with radius {radius}:")
print(f"  Area          = {area:.2f}")
print(f"  Circumference = {circ:.2f}")
```

## 🔁 Part 2 — Loops

A **for loop** repeats code a set number of times.
`range(1, 13)` gives the numbers 1, 2, 3 ... 12.

```python
# Print the 7 times table
for i in range(1, 13):
    result = 7 * i
    print(f"7 × {i:2} = {result}")
```

```python
# 🎯 Try it yourself!
# Change the number below to print a different times table
number = 9

for i in range(1, 13):
    print(f"{number} × {i:2} = {number * i}")
```

```python
# A while loop keeps going as long as a condition is true
# Find the first power of 2 bigger than 1000
n = 1
while n <= 1000:
    n = n * 2

print(f"First power of 2 above 1000 is {n}")
```

## ⚡ Part 3 — Functions

A **function** is a named recipe — define it once with `def`, then call it whenever you need it.
You pass in inputs (called *parameters*) and get back a result.

```python
def is_prime(n):
    """Returns True if n is a prime number."""
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

# Find all primes up to 50
primes = [n for n in range(2, 51) if is_prime(n)]
print("Primes up to 50:")
print(primes)
```

```python
# 🎯 Try it yourself!
# Test whether your favourite numbers are prime
for n in [17, 42, 97, 100, 101]:
    if is_prime(n):
        print(f"{n} is prime ✓")
    else:
        print(f"{n} is not prime")
```

## 📈 Part 4 — Plotting with matplotlib

Python can draw charts! The `matplotlib` library is the most popular one.
Run the cells below to plot some curves.

```python
import matplotlib.pyplot as plt

# Define a function
def square(x):
    return x ** 2

# Generate x and y values
xs = list(range(1, 11))
ys = [square(x) for x in xs]

print("x values:", xs)
print("y = x² : ", ys)
```

```python
# Plot y = x²
plt.figure(figsize=(7, 4))
plt.plot(xs, ys, marker='o', color='purple', linewidth=2)
plt.title("y = x²  (a parabola)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 🎯 Try it yourself — plot multiple curves on the same chart!
xs = list(range(1, 11))

plt.figure(figsize=(7, 4))
plt.plot(xs, [x**1 for x in xs], label='y = x',   color='teal')
plt.plot(xs, [x**2 for x in xs], label='y = x²',  color='purple')
plt.plot(xs, [x**3 for x in xs], label='y = x³',  color='coral')

plt.title("Powers of x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 🌟 Bonus — Fibonacci spiral plot

When you've done the challenges, try running this — it plots the Fibonacci numbers as a bar chart!

```python
import matplotlib.pyplot as plt

# Generate Fibonacci numbers
fibs = []
a, b = 1, 1
for _ in range(12):
    fibs.append(a)
    a, b = b, a + b

plt.figure(figsize=(8, 4))
plt.bar(range(1, 13), fibs, color='mediumpurple', edgecolor='white')
plt.title("Fibonacci Numbers")
plt.xlabel("Position in sequence")
plt.ylabel("Value")
plt.xticks(range(1, 13))
plt.grid(axis='y', alpha=0.3)
plt.show()

print("The numbers:", fibs)
```
